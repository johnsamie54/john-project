```python
#labeling the categorical fetures values
PortNumber={"Port#:1":0,"Port#:2":1, "Port#:3":2 ,"Port#:4":3}
attack={"Attack":0,"Normal":1}
label={"TCP-SYN":0,"Blackhole":1 ,"Diversion":2 ,"Overflow":3,"Normal":4,"PortScan":5}
```

```python
#mapping labeled featres to original dataset
intrusion_detect["Port Number"]=intrusion_detect["Port Number"].map(PortNumber)
intrusion_detect["Binary Label"]=intrusion_detect["Binary Label"].map(attack)
intrusion_detect["Label"]=intrusion_detect["Label"].map(label)
```

```python
#dropping features of based on relationship in the above heatmap
intrusion_detect.drop(["Connection Point","Delta Received Bytes","Delta Sent Bytes"],axis=1,inplace=True)
```

```python

#splitting data into depedent feature and idependent features
x=intrusion_detect.drop("Binary Label",axis=1)
y=intrusion_detect["Binary Label"]
```

```python
#splitting data to train and test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=0)
x_train.shape,x_test.shape
```

```python
#applying artificial neuro_network
#importing various libraries
from keras.wrappers.scikit_learn import KerasClassifier
from keras.models import Sequential
from keras.layers import Dense,LeakyReLU,ELU,PReLU,Dropout,Activation,Embedding,Flatten,BatchNormalization
from keras.activations import relu, sigmoid
```

```python
#creating the model function
def create_model(layers, activation):
    model = Sequential()
    for i, nodes in enumerate(layers):
        if i==0:
            model.add(Dense(nodes,input_dim=x_train.shape[1]))
            model.add(Activation(activation))
            model.add(Dropout(0.3))
        else:
            model.add(Dense(nodes))
            model.add(Activation(activation))
            model.add(Dropout(0.3))
            
    model.add(Dense(units = 1, kernel_initializer= 'glorot_uniform', activation = 'sigmoid')) # Note: no activation beyond this point
    
    model.compile(optimizer='adam', loss='binary_crossentropy',metrics=['accuracy'])
    return model
```

```python
#fitting the function to the model
model = KerasClassifier(build_fn=create_model, verbose=0)
```

```python
#hypertuning the model
from sklearn.model_selection import GridSearchCV
layers = [(20,), (40, 20),(45, 30, 15)]
activations = ['sigmoid', 'relu']
param_grid = dict(layers=layers, activation=activations, batch_size = [128, 256], epochs=[30])
grid = GridSearchCV(estimator=model, param_grid=param_grid,cv=5)
```

```python
grid.fit(x_train,y_train)
```

```python
#finding the best params and best model score
[[grid.best_score_,grid.best_params_]]
```

```python
pred_y = grid.predict(x_test)
y_pred = (pred_y > 0.5)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
#finding erro and accuracy of the model
cm=confusion_matrix(y_pred, y_test)
score = accuracy_score(y_pred, y_test)
print(cm)
print(score)
```
